```python
import yfinance as yf
import pandas_ta as ta
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

# mesures total number of traded asets in a given period.
# Volume is the strength, and conviction behind a price move. 
# It makes trend or reversal more probabel.


df = yf.download(tickers='BTC',start='2024-01-01', end='2024-04-30', interval='1d')

# OBV = On Balance Volume: Volume change
# cumulative total of the volume additions and subtractions forms the OBV line. 
df['OBV'] = ta.obv(df.Close, df.Volume)

df = df[:500]

figure = make_subplots(rows=2, cols=1, subplot_titles=['Price', 'OBV'], shared_xaxes=True)
figure.add_trace(go.Candlestick(x=df.index, open=df['Open'], high=df['High'], low=df['Low'], close=df['Close']), row=1, col=1)
figure.add_trace(go.Scatter(x=df.index, y=df['OBV'], name='OBV'), row=2, col=1)
figure.update_layout(paper_bgcolor="black",  plot_bgcolor="black")  
figure.update_layout(margin_l=10, margin_b=0, margin_r=0, margin_t=0, grid_columns=1,grid_rows=1)
figure.update_yaxes( gridcolor='black')
figure.update_xaxes( gridcolor='black')
figure.update_layout(    xaxis=dict(rangeslider=dict(visible=False)))


figure.show()
```

```python
import yfinance as yf
import pandas_ta as ta
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

df = yf.download(tickers='BTC',start='2024-01-01', end='2024-04-30', interval='1d')
#CMF = Chain Money Flow: mesure the buying and selling pressure
# selling pressure/berish 0 buying pressure/ bullish
df['CMF'] = ta.cmf(df.High, df.Low, df.Close, df.Volume, length=20)
#df = df[1000:1500]

figure = make_subplots(rows=2, cols=1, subplot_titles=['Price', 'CMF'], shared_xaxes=True)
figure.add_trace(go.Candlestick(x=df.index, open=df['Open'], high=df['High'], low=df['Low'], close=df['Close']), row=1, col=1)
figure.add_trace(go.Scatter(x=df.index, y=df['CMF'], name='CMF'), row=2, col=1)
figure.update_layout(paper_bgcolor="black",  plot_bgcolor="black")  
figure.update_layout(margin_l=10, margin_b=0, margin_r=0, margin_t=0, grid_columns=1,grid_rows=1)
figure.update_yaxes( gridcolor='black')
figure.update_xaxes( gridcolor='black')
figure.update_layout(    xaxis=dict(rangeslider=dict(visible=False)))

figure.show()
```
