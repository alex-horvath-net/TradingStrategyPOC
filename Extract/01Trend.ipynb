# Trend indicators


```python
import yfinance as yf # Version: 0.2.38
import pandas as pd   # Version: 2.2.2
import pandas_ta as ta   # Version: 2.2.2
import plotly.graph_objects as go # Version: 5.22.0
from plotly.subplots import make_subplots
import matplotlib.pyplot as pp # Version: 3.9.0

df = yf.download(tickers='USD',start='2023-01-01', end='2024-04-30', interval='1d')
#df.reset_index(inplace=True)
df = df[ df["Volume"]!=0 ] # clean empty volumes
#df = df[:]
df = df.copy()

df["SMA"] = ta.sma(df.Close, length=10) # identifies long term direction (trend, resistance or support leve) by flattening out fluctuations. 
df["EMA"] = ta.ema(df.Close, length=10) # identifis the latest trend, resistance or support leve by flattening out old spices sikes. 
df.ta.adx(high='High', low='Low', close='Close', length=14, append=True) # identifis strength, stability of trend, 0week20moderate40strong60verystrong. 20-25 buy/sell, 50-75 hold, 75-100 exit position. 

close = go.Scatter(x=df.index, y=df["Close"], name='Close',  mode='lines', line=dict(color='blue', width=2),)
candlestick = go.Candlestick( x=df.index, open=df['Open'], high=df['High'], low=df['Low'], close=df['Close'], name="Candlestick")
#signal = go.Scatter(x=df.index, y=df["ADX_14"], mode="markers", marker=dict(size=1, color="moccasin"), name="Signal")

sma = go.Scatter(x=df.index, y=df.SMA, name='SMA',  mode='lines', line=dict(color='green', width=2))
ema = go.Scatter(x=df.index, y=df.EMA, name='EMA',   mode='lines', line=dict(color='yellow', width=2))
adx = go.Scatter(x=df.index, y=df['ADX_14'], name='ADX', yaxis='y2', line=dict(color='red', width=2))


figure = go.Figure(data=[sma, ema, adx])
figure.update_layout(paper_bgcolor="black",  plot_bgcolor="black")  
figure.update_layout(margin_l=10, margin_b=0, margin_r=0, margin_t=0, grid_columns=1,grid_rows=1)
figure.update_layout(yaxis2=dict(domain=[0, 1], anchor='free', overlaying='y',  side='right', title='ADX') )
figure.show()
```
