```python
import yfinance as yf
import pandas_ta as ta
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

# mesures magnitode of price changes

df = yf.download(tickers='BTC',start='2024-01-01', end='2024-04-30', interval='1d')
df = df[ df["Volume"]!=0 ] # clean empty volumes

# volatility means change frequency
# mesures price fluctations per time unit

# BB = Bollinger Bands: 2 std deviations from the MA
# identify overbought and oversold conditions and predict price reversal
df = pd.concat([df, ta.bbands(df.Close, length=14)], axis=1)
df.columns = df.columns[:6].tolist() + ['BB_LOWER', 'BB_MIDDLE', 'BB_UPPER'] + df.columns[9:].tolist()


figure = go.Figure()
figure.add_trace(go.Candlestick( x=df.index, open=df['Open'], high=df['High'], low=df['Low'], close=df['Close'], name="Candlestick"))
figure.add_trace(go.Scatter(x=df.index, y=df['BB_UPPER'], name='BB_UPPER'))
figure.add_trace(go.Scatter(x=df.index, y=df['BB_MIDDLE'], name='BB_MIDDLE'))
figure.add_trace(go.Scatter(x=df.index, y=df['BB_LOWER'], name='BB_LOWER'))


figure.update_layout(paper_bgcolor="black",  plot_bgcolor="black")  
figure.update_layout(margin_l=10, margin_b=0, margin_r=0, margin_t=0, grid_columns=1,grid_rows=1)
figure.update_yaxes( gridcolor='black')
figure.update_xaxes( gridcolor='black')
figure.update_layout(    xaxis=dict(rangeslider=dict(visible=False)))

figure.show()
```

```python
import yfinance as yf
import pandas_ta as ta
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

df = yf.download(tickers='BTC',start='2024-01-01', end='2024-04-30', interval='1d')

# ATR = Average True Range: measure min to max distance
#low atr means low volatility
# stop-los = price +- ATR
 
df['ATR'] = ta.atr(df.High, df.Low, df.Close, length=14)

#df = df[:500]

figure = make_subplots(rows=2, cols=1, subplot_titles=['Price', 'ATR'], shared_xaxes=True)

figure.add_trace(go.Candlestick(x=df.index, open=df['Open'], high=df['High'], low=df['Low'], close=df['Close']), row=1, col=1)
figure.add_trace(go.Scatter(x=df.index, y=df['ATR'], name='ATR'), row=2, col=1)

figure.update_layout(
    xaxis=dict(rangeslider=dict(visible=False))
)

figure.show()
```
